***GPT 4.0 mini Object Identifier***


```python
from openai import OpenAI
import ast

def getobjects(user_input):
  identified_objects = []
  messages=[
      {"role": "system", "content": "Extract simple and drawable objects from the prompt. Avoid backgrounds/setting/place. The prompt will have the following format, \"Prompt: \"A cat sitting on a mat.\" Return in the objects in a list like as follows \"Cat\", \"Mat\""},
    ]

  client = OpenAI()

  messages.append({"role":"user","content":user_input})

  completion = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
  )
  identified_objects = completion.choices[0].message.content
  identified_objects = ast.literal_eval(identified_objects)

  print(completion.choices[0].message.content)
  return identified_objects
```

***GPT 4o mini Prompt Generator***


```python
def get_edited_prompts(user_input,identified_objects):

    edited_prompts = []
    print("Original Prompt:", user_input)
    
    for obj in identified_objects:
        print("Doing for ojb = ",obj)
        
        instruction = f"""You are given a prompt and an object. 
        The prompt describes a scene in which multiple objects are interacting with each other.
        The object you are provided with is one of the objects in the prompt.
        You are to make a simpler prompt from the original one that focuses more on given object and not at the other objects in the original prompt.
        Give a very simple prompt. keep the simpler prompt short"""
        
        prompt = f"""Prompt: {user_input}
                object: {obj}"""

        messages=[
            {"role": "system", "content": instruction},
        ]

        client = OpenAI()

        messages.append({"role":"user","content":prompt})

        completion = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages
        )
        edited_prompts.append(completion.choices[0].message.content)

    for new_prompt in edited_prompts:
        print(new_prompt)

    return edited_prompts



```

```python
Original_Prompt = input("Enter your prompt")

# Step 1: Identify objects in the prompt
identified_objects = getobjects(Original_Prompt)


# Step 2: Generate edited prompts for each identified object
edited_prompts = get_edited_prompts(Original_Prompt,identified_objects)




# Print arrays to verify
print("Objects Array:", identified_objects)
print("Edited Prompts Array:", edited_prompts)
```
